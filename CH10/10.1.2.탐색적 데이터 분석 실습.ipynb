```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(color_codes=True)
#matplotlib inline

df = pd.read_csv('../datasets/hotel_bookings.csv')
```

```python
df.head() # 데이터 샘플 확인
```

```python
df.info() # 각 칼럼의 속성 및 결측치 확인
```

```python
df.describe() # 각 칼럼의 통계치 확인 
```

```python
df.skew() # 각 칼럼의 왜도 확인
```

```python
df.kurtosis() # 각 칼럼의 첨도 확인
```

```python
# 특정 변수 분포 시각화
sns.distplot(df['lead_time']) # lead_time: 예약 날짜로부터 투숙 날짜까지의 일수 차이
```

```python
# 호텔 구분에 따른 lead_time 분포 차이 시각화
# violinplot: 분포를 효과적으로 표현
sns.violinplot(x='hotel', y='lead_time', data=df, inner=None, color='.8')
# stripplot: 각 관측치의 위치를 직관적으로 표현
sns.stripplot(x='hotel', y='lead_time', data=df, size=1)
```

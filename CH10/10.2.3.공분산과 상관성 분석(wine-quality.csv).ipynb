```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set(color_codes=True)
#matplotlib inline

df = pd.read_csv('../datasets/wine-quality.csv')
```

```python
df.head() # 데이터 샘플 확인
```

```python
# 산점도 행렬 시각화
sns.set(font_scale=1.1)
sns.set_style('ticks') # 축 눈금 설정
sns.pairplot(df, diag_kind='kde') # kde:상관계수가 1이면 분포로 표시
plt.show()
```

```python
df.cov() # 공분산 확인
```

```python
df.corr(method='pearson') # 피어슨 상관계수 확인
```

```python
# 히트맵 시각화
sns.heatmap(df.corr(), cmap='viridis')
```

```python
# clustermap 히트맵 시각화 - clustermap은 기존 히트맵에 상관성이 강한 변수들끼리 묶어 표현
sns.clustermap(df.corr(), annot=True, cmap='RdYlBu_r', vmin=-1, vmax=1)
```

```python
# 중복 제거 히트맵 시각화

# triu(Upper Triangular Matrix): 매트릭스의 우측 상단을 모두 True인 1로, 하단을 False인 0으로 변환 
np.triu(np.ones_like(df.corr()))  

# True/False mask 배열로 변환
mask = np.triu(np.ones_like(df.corr(), dtype=np.bool))

# 히트맵 그래프 생성
fig, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='RdYlBu_r', cbar=True)
ax.set_title('Wine Quality Correlation', pad=15)
```

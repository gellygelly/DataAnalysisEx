```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv('../datasets/50_Startups.csv')

df.head()
```

```python
# 기본 산점도 시각화
plt.scatter(df['R&D Spend'], df['Profit'], s=50, alpha=0.4) # x=R&D spend, y=Profit
plt.show()
```

```python
# 산점도 회귀선 추가
ax = sns.lmplot(x='R&D Spend', y='Profit', data=df)
```

```python
# 네 가지 요소의 정보를 포함한 산점도 시각화

plt.scatter(df['R&D Spend'], # x
            df['Profit'], # y
            df['Marketing Spend']*0.001, # 버블의 크기
            c=df['Administration'], # 버블 색상
              alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```
